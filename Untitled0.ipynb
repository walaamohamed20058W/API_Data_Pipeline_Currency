```python
import requests
import pandas as pd
from datetime import datetime


url = "https://open.er-api.com/v6/latest/USD"

response = requests.get(url)
data = response.json()


print("Available keys:", data.keys())


rates = data["rates"]


df = pd.DataFrame(rates.items(), columns=["Currency", "Rate_to_USD"])
df["Date"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")


df.to_csv("currency_rates.csv", index=False)
df.to_excel("currency_rates.xlsx", index=False)

print("✅ Data pipeline finished! Files saved successfully.")
df.head()
```
